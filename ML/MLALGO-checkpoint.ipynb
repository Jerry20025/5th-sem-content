```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\HP\\Documents\\5th sem\\ML\\MLALGO\\income.csv")
df
```

```python
plt.scatter(df['Age'],df['Income'])
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['Age','Income']] )
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['Income']])
df['Income']=scaler.transform(df[['Income']])
scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
df
```

```python
sse=[]
k_range=range(1,10)
for i in k_range:
    km=KMeans(n_clusters=i)
    km.fit(df[['Age',"Income"]])
    sse.append(km.inertia_)
    
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Square Error')
plt.plot(k_range,sse)
```

```python
km=KMeans(n_clusters=4)
y_predicted=km.fit_predict(df[['Age','Income']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==4]

plt.scatter(df1.Age, df1.Income,color='red',label='cluster1')
plt.scatter(df2.Age, df2.Income,color='green',label='cluster2')
plt.scatter(df3.Age, df3.Income,color='black',label='cluster3')
plt.scatter(df4.Age, df4.Income,color='purple',label='cluster4')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], color='blue' ,marker='*', label='centroid')
plt.legend()
km.cluster_centers_
```
